```python
import pandas
import matplotlib
import numpy
import matplotlib.pyplot as plt
from scipy import stats

df = pandas.read_csv("time_data.csv")
print(df)


df['Time'].plot(kind='bar')

df12 = pandas.DataFrame(data={
    'df1': df['Time'],
})
df12.plot.kde()
plt.show()

d1 = df12['df1']
print(stats.kstest(d1, 'norm', (d1.mean(), d1.std()), N=len(d1)))
```
